```python
from sklearn.preprocessing import scale
from sklearn.utils import shuffle
```

```python
df_shuffled = shuffle(df, random_state=123)
X = scale(df_shuffled[df_shuffled.columns[:-1]])
y = df_shuffled["cnt"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_reg = LinearRegression()
```

```python
linear_reg.fit(X, y)
```

```python
from sklearn.linear_model import Lasso, Ridge
```

```python
lasso_regressor = Lasso(alpha=50)
lasso_regressor.fit(X, y)
```

```python
ridge = Ridge(alpha=1)
ridge.fit(X, y)
```

```python
list(zip(lasso_regressor.coef_))
```

```python
list(zip(ridge.coef_))
```

```python
alphas = np.arange(1, 500, 50)
coefs_lasso = np.zeros((alphas.shape[0], X.shape[1]))
coefs_ridge = np.zeros((alphas.shape[0], X.shape[1]))
```

```python
for i, a in enumerate(alphas):
    
    regrL = Lasso(alpha=a)
    regrL.fit(X, y)
    weights=regrL.coef_
    for j, weight in enumerate(weights):
        coefs_lasso[i,j]=weight
coefs_lasso
```

```python
for i, a in enumerate(alphas):
    
    regrR = Ridge(alpha=a)
    regrR.fit(X, y)
    weights=regrR.coef_
    for j, weight in enumerate(weights):
        coefs_ridge[i,j]=weight
coefs_ridge
```

```python
plt.figure(figsize=(8, 5))
for coef, feature in zip(coefs_lasso.T, df.columns):
    plt.plot(alphas, coef, label=feature, color=np.random.rand(3))
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("alpha")
plt.ylabel("feature weight")
plt.title("Lasso")

plt.figure(figsize=(8, 5))
for coef, feature in zip(coefs_ridge.T, df.columns):
    plt.plot(alphas, coef, label=feature, color=np.random.rand(3))
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("alpha")
plt.ylabel("feature weight")
plt.title("Ridge")
```

```python
from sklearn.linear_model import LassoCV
```

```python
alphas = np.arange(1, 100, 5)

model = LassoCV(alphas=alphas, cv=3).fit(X, y)


plt.plot(model.alphas_, model.mse_path_.mean(axis=1), 
         label='Average across the folds')
plt.grid(True)
plt.legend(loc='upper left')

print("alpha=", model.alpha_)

regrL = Lasso(alpha=model.alpha_)

# Train the model using the sets
regrL.fit(X, y)

# The coefficients
weights=regrL.coef_

print('\nFeature:  Weight:')
      
for feature, coeff in zip(df.columns,regrL.coef_.T):
    print(feature, coeff)

model.mse_path_
```

```python
index_mins = model.mse_path_.argmin(axis=0)
print("alphas_min_columns=", model.alphas_[index_mins])
```

```python
for i in range(3):
    plt.plot(model.alphas_, model.mse_path_[:,i], label='Column'+ str(i+1))
    plt.grid(True)
    plt.legend(loc='upper left')
    plt.show()
```
